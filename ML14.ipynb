1) What is Dimensionality Reduction? Why is it important in machine
learning?
->
Dimensionality reduction is the process of reducing the number of input variables (features) in a dataset while retaining its essential information and patterns. It converts high-dimensional data into a lower-dimensional space, addressing the "curse of dimensionality" to improve machine learning model performance, reduce computation time, and enhance data visualization.

**Why Dimensionality Reduction is Important in Machine Learning**
Mitigates the Curse of Dimensionality: As features increase, data becomes sparse, making it hard for models to find patterns; reducing dimensions helps models train effectively.
Improves Model Performance: Removing irrelevant, noisy, or redundant features prevents overfitting and improves the accuracy of machine learning models.
Speeds Up Training Time: Models require less computational power and memory to process, leading to faster training and inference.
Enhances Visualization: It enables the visualization of complex, high-dimensional data in 2D or 3D, aiding in exploratory data analysis and pattern recognition.
Reduces Storage Space: Lower-dimensional data requires less storage space.

3)What is clustering in unsupervised learning? Mention three popular
clustering algorithms.
->
Clustering is an unsupervised machine learning technique that automatically groups unlabeled data points into clusters based on shared characteristics or similarity measures. The goal is to maximize intra-cluster similarity (points inside a group are alike) and minimize inter-cluster similarity.

**Here are three popular clustering algorithms:**
-K-Means Clustering: Partitions data into
 distinct, non-overlapping clusters by minimizing the distance between data points and their respective cluster centroid.
-Hierarchical Clustering: Builds a tree-like hierarchy of clusters (dendrogram) by iteratively merging (agglomerative) or splitting (divisive) clusters.
-DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Groups together points that are in close proximity (high-density regions), making it effective for identifying clusters of arbitrary shapes and detecting noise/outliers.

4)Explain the concept of anomaly detection and its significance.
->
Anomaly detection is the process of identifying data points, events, or observations that deviate significantly from a dataset's established normal behavior, often referred to as outliers, exceptions, or novelties. Utilizing AI and machine learning, it automatically spots irregularitiesâ€”such as system failures, fraudulent transactions, or security breachesâ€”that do not conform to expected patterns.

**Core Concept of Anomaly Detection**

Defining "Normal": It begins by analyzing historical data to establish a baseline of normal behavior.
Identifying Outliers: It scans new data to find points that fall outside this established norm.
Types of Anomalies:
Point Anomalies: A single, isolated data point is far from the rest.
Contextual Anomalies: Data is normal in general but abnormal in a specific context (e.g., high spending on a weekend is normal, but on a Monday at 3 AM it is not).
Collective Anomalies: A set of data points is anomalous, even if individual points are not.

**Significance and Importance**

Anomaly detection is critical because it identifies potential issues before they become critical failures, reducing risks and costs.

Cybersecurity: Detects, unauthorized access, intrusions, or unusual network behavior.
Finance: Identifies fraudulent credit card transactions, money laundering, or suspicious trading patterns.
Manufacturing: Monitors machine sensor data to predict equipment failure before downtime occurs.
IT Operations: Reduces "alert fatigue" by spotting performance bottlenecks or bugs in complex systems.
Healthcare: Detects irregular patient conditions or billing fraud.
It automates what was once manual data analysis, enabling proactive, data-driven decision-making.

->
Here are **three common types of anomaly detection techniques** with brief descriptions:

---

## ðŸ”¹ 1. Statistical Methods

**Description:**
These techniques assume the data follows a known statistical distribution (e.g., normal distribution). Points that significantly deviate from the expected distribution are flagged as anomalies.

**How it works:**

* Calculate mean and standard deviation
* Use Z-score or probability thresholds
* If a value lies far from the mean â†’ anomaly

**Example methods:**

* Z-score
* Gaussian models
* IQR (Interquartile Range)

**Best for:**
Simple datasets with known distributions.

---

## ðŸ”¹ 2. Distance-Based Methods

**Description:**
These methods detect anomalies based on how far a data point is from others in the dataset.

**How it works:**

* Compute distances between points
* Points far from clusters are considered anomalies

**Example methods:**

* K-Nearest Neighbors (KNN)
* DBSCAN

**Best for:**
Multidimensional datasets where anomalies are isolated points.

---

## ðŸ”¹ 3. Density-Based Methods

**Description:**
These techniques identify anomalies as points that lie in **low-density regions** compared to their neighbors.

**How it works:**

* Measure local density around each point
* Points with significantly lower density are anomalies

**Example methods:**

* Local Outlier Factor (LOF)
* Isolation Forest

**Best for:**
Complex datasets where anomalies are not just far away but located in sparse regions.

## ðŸ“Š Difference Between Seasonality and Cyclic Behavior in Time Series

Both **seasonality** and **cyclic behavior** describe repeating patterns in time series data, but they differ in **regularity and time period**.

---

## ðŸ”¹ Seasonality

**Definition:**
Seasonality refers to patterns that repeat at **fixed and known time intervals**.

**Key characteristics:**

* Regular and predictable
* Fixed frequency
* Short-term repetition
* Calendar-based

**Examples:**

* Ice cream sales increase every summer
* Retail sales spike every December
* Higher traffic during weekdays

ðŸ‘‰ The timing is consistent and known in advance.

---

## ðŸ”¹ Cyclic Behavior

**Definition:**
Cyclic behavior refers to **long-term oscillations** in data that occur over **irregular or variable periods**.

**Key characteristics:**

* Not strictly periodic
* Duration varies
* Often linked to economic/business cycles
* Longer-term patterns

**Examples:**

* Economic boom and recession cycles
* Long-term cryptocurrency bull and bear markets
* Business growth cycles




````python
#8) Write Python code to perform K-means clustering on a sample dataset.
#(Include your Python code and output in the code box below.)
# K-means Clustering on a Sample Dataset

# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 2: Create sample data
X, y_true = make_blobs(n_samples=200, centers=3, random_state=42)

# Step 3: Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)

# Step 4: Get results
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Step 5: Print output
print("Cluster Centers:")
print(centers)

print("\nFirst 10 Cluster Labels:")
print(labels[:10])

# Step 6: Visualize clusters
plt.figure(figsize=(6,5))
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], marker='X', s=200)
plt.title("K-Means Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
````

9)What is inheritance in OOP? Provide a simple example in Python.
->
Inheritance is a core principle of Object-Oriented Programming (OOP) that allows a new class to adopt the properties and behaviors (methods and attributes) of an existing class [1]. This mechanism promotes code reusability and establishes a natural hierarchy between classes. The original class is called the parent (or superclass), and the new class is called the child (or subclass) [1, 2].

e.g.
# Parent class
class Animal:
    def speak(self):
        print("Animal makes a sound")

# Child class (inherits from Animal)
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Create object of child class
d = Dog()

# Access inherited method
d.speak()   # from Animal

# Access own method
d.bark()    # from Dog

## ðŸ Simple Python Example (Rolling Z-score)

```python
import pandas as pd
import numpy as np

# sample time series
ts = pd.Series([10,11,12,13,50,12,11,10])

# rolling statistics
mean = ts.rolling(window=3).mean()
std = ts.rolling(window=3).std()

z_score = (ts - mean) / std

# anomaly flag
anomalies = ts[np.abs(z_score) > 2]

print(anomalies)
```
